```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x_line = np.linspace(0,10,100)
y_line = np.sin(x_line)
```

```python
#gerar o grafico de linhas
fig = plt.figure(figsize=(10,5))
fig.set_facecolor('orange')

plt.plot(x_line, y_line)
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.title("Gráfico de Linhas")
plt.show()
```

Subplot ou Subgráficos

```python
fig, axs = plt.subplots(2,2, figsize = (10,5), layout='constrained')

fig.set_facecolor('orange')
axs[0,0].set_facecolor('blue')
axs[0,1].set_facecolor('red')
axs[1,0].set_facecolor('yellow')
axs[1,1].set_facecolor('purple')

axs[0,0].set_title('Titulo 1')
axs[0,1].set_title('Titulo 2')
axs[1,0].set_title('Titulo 3')
axs[1,1].set_title('Titulo 4')
```

```python
#Gráfico de barras ou Bar

x = ['Chandler', 'Joey', 'Monica', 'Phoebe']

y = [26,10,15,28]
```

```python
plt.style.use('ggplot')

fig, ax = plt.subplots(figsize=(5,3), layout = 'constrained')
ax.set_xlabel('Eixo X', fontsize = 9)
ax.set_ylabel('Eixo Y', fontsize = 9)
ax.set_title('Título do Gráfico', fontsize = 12)
plt.bar(x,y)
```

## Seaborn

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df =sns.load_dataset('tips')
df.head(5)
```

```python
sns.barplot(x='day', y='tip', data=df)
```

```python
#relação entre o valor da gorjeta e valor da conta
sns.scatterplot(x='tip', y='total_bill', data=df)
plt.title('Valor da Gorjeta X Valor de conta')
```

```python
#comparação de gorjetas por gênero
sns.set_style('dark')
sns.boxplot(x='sex', y='tip', data=df)
plt.title('Gorjeta por Gênero')
```
